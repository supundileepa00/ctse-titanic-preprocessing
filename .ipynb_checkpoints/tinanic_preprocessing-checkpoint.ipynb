#### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv("Titanic Dataset.csv")
data_frame.head()
```

```python
# Create a histogram of the Age column
data_frame['Age'].hist(bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Age')
plt.show()

# Create a histogram of the Fare column
data_frame['Fare'].hist(bins=20)
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Distribution of Fare')
plt.show()

```

```python
# Create box plots of the Age and Fare columns grouped by Survived
data_frame.boxplot(column=['Age'], by='Survived')
plt.title('Age by Survived')
plt.suptitle('')
plt.ylabel('Age')
plt.show()

data_frame.boxplot(column=['Fare'], by='Survived')
plt.title('Fare by Survived')
plt.suptitle('')
plt.ylabel('Fare')
plt.show()

```

```python
# Create a bar plot of the Pclass column
data_frame['Pclass'].value_counts().plot(kind='bar')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.title('Distribution of Pclass')
plt.show()

# Create a bar plot of the Sex column
data_frame['Sex'].value_counts().plot(kind='bar')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Distribution of Sex')
plt.show()

# Create a bar plot of the Embarked column
data_frame['Embarked'].value_counts().plot(kind='bar')
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Distribution of Embarked')
plt.show()

# Create bar plots of the Pclass, Sex, and Embarked columns grouped by Survived
pd.crosstab(data_frame['Pclass'], data_frame['Survived']).plot(kind='bar', stacked=True)
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.title('Survival Rate by Pclass')
plt.show()


pd.crosstab(data_frame['Sex'], data_frame['Survived']).plot(kind='bar', stacked=True)
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.title('Survival Rate by Sex')
plt.show()

pd.crosstab(data_frame['Sex'], data_frame['Survived']).plot(kind='bar', stacked=True)
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Survival Rate by Sex')
plt.show()

pd.crosstab(data_frame['Embarked'], data_frame['Survived']).plot(kind='bar', stacked=True)
plt.xlabel('Embarked')
plt.ylabel('Count')
plt.title('Survival Rate by Embarked')
plt.show()
```

```python
# remove unnecessary columns
data_frame.drop(columns=[ "PassengerId","Name","Ticket","Cabin"], inplace=True)
print(data_frame)
```

```python
# Impute missing values in the Age column using median imputation
median_age = data_frame['Age'].median()
data_frame['Age'].fillna(median_age, inplace=True)


# Impute missing values in the Fare column using median imputation
median_fare = data_frame['Fare'].median()
data_frame['Fare'].fillna(median_fare, inplace=True)

# # Divide the Fare column into quantiles
data_frame['FareQuantile'] = pd.qcut(data_frame['Fare'], q=4, labels=['1st', '2nd', '3rd', '4th'])
data_frame.drop('Fare', axis=1, inplace=True)

# Create a binary variable for the Sex column
data_frame['Sex'] = data_frame['Sex'].map({'female': 0, 'male': 1})

# One-hot encode the Pclass, Sex, SibSp, Parch,and Embarked columns
data_frame = pd.get_dummies(data_frame, columns=['Pclass', 'Embarked'])

# Print the resulting dataframe
print(data_frame.head())
```
